```python
from pytrends.request import TrendReq
import pandas as pd

class GoogleTrends:
    def __init__(self):
        self.pytrend = TrendReq(hl='en-US', tz=360)
        
    def build_payload(self, kw_list):
        self.pytrend.build_payload(kw_list=kw_list)
        
    def get_interest_by_region(self):
        return self.pytrend.interest_by_region(resolution='COUNTRY', inc_low_vol=False)
        
    def get_top_charts(self, year, geo):
        return self.pytrend.top_charts(year, hl='en-US', tz=300, geo=geo)
    
    def get_historical_interest(self, 
                                kw_list,
                                year_start,
                                month_start,
                                day_start,
                                hour_start,
                                year_end,
                                month_end,
                                day_end,
                                hour_end,
                                cat,
                                geo,
                                gprop,
                                sleep
                               ):
        return self.pytrend.get_historical_interest(
            kw_list, 
            year_start=2018,
            month_start=1,
            day_start=1,
            hour_start=0,
            year_end=2018,
            month_end=2,
            day_end=1,
            hour_end=0,
            cat=0,
            geo='',
            gprop='',
            sleep=0
        )
    
    def get_related_queries(self):
        return self.pytrend.related_queries().values()
```

```python
kw_list = [
    'Flood'
]
year_start = 2015
month_start=1,
day_start=1,
hour_start=0,
year_end=2021,
month_end=7,
day_end=16,
hour_end=0,
cat=0,
geo='US',
gprop='',
sleep=10

gt = GoogleTrends()
historical_interest = gt.get_historical_interest(
    kw_list=kw_list,
    year_start=year_start,
    month_start=month_start,
    day_start=day_start,
    hour_start=hour_start,
    year_end=year_end,
    month_end=month_end,
    day_end=day_end,
    hour_end=hour_end,
    cat=cat,
    geo=geo,
    gprop=gprop,
    sleep=sleep
)
historical_interest
```
